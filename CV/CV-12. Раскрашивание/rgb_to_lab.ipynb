```python
import skimage
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'png'
pio.templates.default = 'plotly_dark'
```

```python
img = skimage.io.imread('hawaii.jpg')
px.imshow(img).show()
```

```python
img_lab = skimage.color.rgb2lab(img)
```

```python
px.imshow(img_lab[..., 0], color_continuous_scale='Greys_r').show()
```

```python
# Adjust brightness
img_lab[..., 0] = img_lab[..., 0] + 30
img_rgb_1 = skimage.color.lab2rgb(img_lab)
px.imshow(
    img_rgb_1,
    title='<b>  *<i>L</i>-,   30</b>'
).update_layout(
    xaxis_visible=False, 
    yaxis_visible=False,
    title_x=0.5
).write_image('lab_experiments/adjusted_brightness.png')
# Subtract *a channel (green and red channel)
img_lab = skimage.color.rgb2lab(img)
img_lab[..., 1] = 0
img_rgb_2 = skimage.color.lab2rgb(img_lab)
px.imshow(
    img_rgb_2,
    title='<b>  *<i>a</i>-,  0</b>'
).update_layout(
    xaxis_visible=False, 
    yaxis_visible=False,
    title_x=0.5
).write_image('lab_experiments/subtracted_a_channel.png')
# Subtract *b channel (blue and yellow channel)
img_lab = skimage.color.rgb2lab(img)
img_lab[..., 2] = 0
img_rgb_3 = skimage.color.lab2rgb(img_lab)
px.imshow(
    img_rgb_3,
    title='<b>  *<i>b</i>-,  0</b>'
).update_layout(
    xaxis_visible=False, 
    yaxis_visible=False,
    title_x=0.5
).write_image('lab_experiments/subtracted_b_channel.png')
```
